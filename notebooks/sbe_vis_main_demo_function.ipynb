This notebook is to demo the demo display function. The main function takes a model object and a generator object.


What the function does is take the datapoints created by the generator, and train (if the model is not pre-trained) the model on the first N data points, where N is the size of the sliding window input.

It will then take steps of N/4 to move over the datapoints and predict the next N datapoints, measure the input metric, and display it in a plot.
